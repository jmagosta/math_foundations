## Question 7

A vector orthogonal to $(\sqrt 3, \sqrt 3, \sqrt 3)/3 $ is

$(\sqrt 2, - \sqrt 2, 0)/2$  Any rearrangement of elements also works.

With orthonormal vectors, the gram matrix is the identity, so the projection operator reduces to $A A^T$

```python
# Better to make these all unit vectors:
s3 = math.sqrt(3) /3
s2 = math.sqrt(2) /2
x = np.array([1,0,0])

# There are three cases, for the orthogonal vector
A = np.array([[s3,-s2], [s3, 0 ], [s3, s2]])
A
```

```python
A.T @ x
```

```python
# The inverse gram matrix:
la. inv(A.T @ A)
```

```python
# One solution
A @ A.T @ x
```

```python
# Another solution
A = np.array([[s3,s2], [s3, -s2 ], [s3, 0]])
A @ A.T @ x
```

```python
# A third solution
A = np.array([[s3, 0], [s3, -s2 ], [s3, s2]])
A @ A.T @ x
```

```python
# Another orthogonal case
a1 = np.array([1,1,1])/ math.sqrt(3)
a2 = np.array([1,1,-2])/math.sqrt(6)
A = np.vstack([a1,a2]).T
A.T @ A
```

```python
A @ A.T @ x
```

## Question 9

Reverse diagonal matrix determinant.

```python
import numpy as np

def rev_diag(n):
  m = np.zeros((n,n))
  m[0,0] = 0.7     # show that adding an off-diagonal element doesn't change anything.
  for k in range(n):
    m[k, n-k-1] = 1
  return m

def odd(n):
  return(n % 2 == 1)

for i in range(1, 10):
  print(i, '\t', np.linalg.det(rev_diag(i)))
```
