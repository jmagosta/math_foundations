# We'll use two columns from the Titanic dataset to visualize how one variable depends on another.

We can show this in a few ways. Specifically we want to show a _conditional_ dependence of one variable on another.

```python
# Load the standard data manipulation libraries
import numpy as np
import pandas as pd

# There are 22 datasets available from the Seaborn statistical plotting package
# Seaborn also has plotting tools specific to statistical analysis that we can use.
import seaborn as sns
```

```python
# The titanic dataset lists passengers by row. First, lets just look at the marginals by using Colab to plot histograms
# for some for the random variables, and select one - age for example.
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Using Colab's built-in recommended plots

from matplotlib import pyplot as plt
titanic['age'].plot(kind='hist', bins=20, title='age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Let's break out age by class in a grouped barplot using seaborn's catplot

f, ax = plt.subplots(figsize=(7, 5))
ax. set_title("Age Histogram by Class")
sns.despine(f)

sns.histplot(
    titanic,
    x="age", hue="class",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5
)
# sns.title("Age Histogram by Class")
```

```python
# Let's reverse that, to see how class predicts age instead.
# Draw a nested boxplot to show bills by day and time
sns.set_theme(style="ticks", palette="pastel")

sns.boxplot(x="class", y="age", hue='class',
             palette=["m", "g", "r"],
            data=titanic)
sns.despine(offset=10, trim=True)
```
